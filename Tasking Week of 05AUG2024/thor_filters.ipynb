### Listings Filter

```python
def apply_filters(dataframe, filters = None):
    # Initialize columns for filter identification
    dataframe['oden_filter_name'] = None
    all_conditions = []

    for filter_dict in filters:
        filter_name = filter_dict['FilterName']
        filter_match = filter_dict.get('FilterMatch', {})  # Get FilterMatch dict, default to empty dict if not present
        condition = pd.Series([True] * len(dataframe))
        filter_match_mappings = {}

        for criteria in filter_dict['Filter']:
            field = criteria['Field']
            filter_type = criteria['FilterType']
            filter_match_criteria = criteria.get('FilterMatch')

            if filter_match_criteria:
                filter_match_mappings[field] = filter_match_criteria

            if filter_type == 'range':
                condition &= (dataframe[field].fillna(0).astype(int) >= criteria['MinValue']) & (dataframe[field].fillna(0).astype(int) <= criteria['MaxValue'])
            elif filter_type == '=':
                if isinstance(criteria['Value'], str):
                    condition &= (dataframe[field].fillna('').str.lower() == criteria['Value'].lower())
                else:
                    condition &= (dataframe[field].fillna(0).astype(int) == criteria['Value'])
            elif filter_type == '<=':
                condition &= (dataframe[field].fillna(0).astype(int) <= criteria['Value'])
            elif filter_type == '>=':
                condition &= (dataframe[field].fillna(0).astype(int) >= criteria['Value'])
            elif filter_type == '>':
                condition &= (dataframe[field].fillna(0).astype(int) > criteria['Value'])
            elif filter_type == '<':
                condition &= (dataframe[field].fillna(0).astype(int) < criteria['Value'])
            elif filter_type == 'in':

                if all(isinstance(val, str) for val in criteria['Value']):
                    lower_values = [val.lower() for val in criteria['Value']]
                    condition &= dataframe[field].fillna('').str.lower().isin(lower_values)
                else:
                    condition &= dataframe[field].fillna(0).astype(int).isin(criteria['Value'])
            elif filter_type == 'contains':
                if isinstance(criteria['Value'], list):
                    # Check if any of the list items is a substring of the DataFrame cell
                    condition &= dataframe[field].apply(lambda x: any(val.lower() in str(x).lower() for val in criteria['Value']))
                else:
                    # Check if the single string is a substring of the DataFrame cell
                    condition &= dataframe[field].fillna('').str.lower().str.contains(criteria['Value'].lower())

        for field, new_col in filter_match_mappings.items():
                    dataframe.loc[condition, new_col] = dataframe.loc[condition, field]
        
        all_conditions.append(condition)
        dataframe.loc[condition, 'oden_filter_name'] = filter_name
        
        # Assign FilterMatch values to new columns for matched rows
        for key, value in filter_match.items():
            dataframe.loc[condition, key] = value

    final_condition = pd.Series([False] * len(dataframe))
    for condition in all_conditions:
        final_condition |= condition


    return dataframe#[final_condition]
```
